```python
import pandas as pd

books = pd.read_excel("Data Excel/Data.xlsx")
```

```python
books
```

```python
books['Category']
```

```python
books['Category'] = books['Category'].str.replace('\n', '', regex=False)
books['Category']
```

```python
books
```

```python
books[[ 'Author','Price']].value_counts(normalize=True)
```

```python
books['Price'] = books['Price'].str.replace('£', '', regex=False).astype(float)
```

```python
books['Total'] = books['Price'] * books['Quantity']
books['Total']
```

```python
Total_monee = books.groupby('Author')['Total'].sum()
Total_monee
```

```python
num_books = books.groupby('Author').size()
```

```python
summ = pd.DataFrame({
    'Total Revenue' : Total_monee,
    'Total Books' : num_books
})
# summ.reset_index(inplace=True)
summ
```

```python
summ.to_excel('Output Excels\\Authors_Summary.xlsx')
```
